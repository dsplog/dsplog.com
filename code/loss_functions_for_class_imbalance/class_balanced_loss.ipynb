## Class Balanced Weights

```python
import numpy as np

def get_cb_weights(samples_per_cls, beta):
    """
    Calculates Class-Balanced weights.
    Handle beta=1 (Inverse Frequency) and beta=0 (No weighting) specifically.
    """
    samples_per_cls = np.array(samples_per_cls)
    num_classes = len(samples_per_cls)

    if beta == 0:
        # Case: No weighting (Uniform weights)
        weights = np.ones(num_classes)
    elif beta == 1.0:
        # Case: Inverse Class Frequency (En = n)
        weights = 1.0 / samples_per_cls
    else:
        # Case: Class-Balanced Loss formula
        effective_num = (1.0 - np.power(beta, samples_per_cls)) / (1.0 - beta)
        weights = 1.0 / effective_num

    # Normalization: sum(weights) = num_classes
    # This keeps the loss scale consistent across different beta values
    weights = weights / np.sum(weights) * num_classes
    return weights

# --- Analysis for Long-Tailed Data ---
samples = [5000, 1000, 500, 100] # [Majority, ..., Minority]
betas = [0, 0.99, 0.9999, 1.0]

print(f"{'Beta':<10} | {'Strategy':<20} | {'Weights per Class (Normalized)':<40}")
print("-" * 85)

for b in betas:
    strategy = "Uniform" if b == 0 else ("Inv-Freq" if b == 1.0 else "Class-Balanced")
    w = get_cb_weights(samples, b)
    
    formatted_w = [round(x, 4) for x in w]
    print(f"{b:<10} | {strategy:<20} | {str(formatted_w):<40}")
```
