### Linear Regression

#### Synthetic Training Data

```python
import numpy as np
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Number of training examples
m = 1000

# Generate 100 random x values
X = np.random.rand(1,m) * 10  # values between 0 and 10

# Define a true linear relationship: y = 2.5x + 5 + some noise
noise = np.random.randn(1,m) * 2  # Gaussian noise
y = 2.5 * X + 5 + noise

# Plot it
plt.scatter(X, y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Synthetic Linear Data")
plt.show()
```

#### Training Loop

```python
# learning rate and number of iterations
alpha = 0.01 
n_iter = 2000

# initializing weights and bias
w = np.random.randn(1,1)
b = np.random.randn(1,1)

# Storing losses and weights/bias for plotting
losses = []
weights = []
biases = []

# Gradient descent
for _ in range(n_iter) : 
    # Forward pass: compute predictions
    z = w.T @ X + b

    # Compute the loss (mean squared error)
    loss = np.sum((z-y)**2)/m
    losses.append(loss)

    # Backward pass: compute gradients   
    dw = (1/m)*X @ (z-y).T
    db = (1/m)*np.ones((1,m)) @ (z-y).T

    # Update weights and bias    
    w -= alpha*dw
    b -= alpha*db

    # Store weights and bias for plotting
    weights.append(w[0,0])
    biases.append(b[0,0])
```

#### Visualizing the loss and weight/bias updates

```python
# Plotting the loss
plt.semilogy(losses)
plt.xlabel("Iterations")
plt.ylabel("Loss")
plt.title("Loss over Iterations")
plt.show()

# Plotting the weights and bias
plt.plot(weights, label='Weight')
plt.plot(biases, label='Bias')
plt.legend()
plt.xlabel("Iterations")
plt.ylabel("Value")
plt.title("Weight and Bias over Iterations")
plt.show()

# Final parameters
print(f"Final weight: {w.item()}")
print(f"Final bias: {b.item()}")
```

#### Final Results

```python
# Final predictions
predictions = w.T @ X + b

# Plotting the final predictions against the true values
plt.scatter(X, y, label='True Values')
plt.scatter(X, predictions, label='Predictions', color='red')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Final Predictions vs True Values")
plt.legend()
plt.show()
```
