## SoftPlus and ReLU activation

```python
import numpy as np
import matplotlib.pyplot as plt

# Activation functions
def relu(x):
    return np.maximum(0, x)

def softplus(x):
    return np.log1p(np.exp(x))   # Numerically stable version

# Input range
x = np.linspace(-6, 6, 500)

# Compute outputs
y_relu = relu(x)
y_softplus = softplus(x)

# Plot
plt.figure(figsize=(7,5))
plt.plot(x, y_relu, label='ReLU', linewidth=2)
plt.plot(x, y_softplus, label='Softplus', linewidth=2)

plt.xlabel('x')
plt.ylabel('Activation')
plt.title('Softplus vs ReLU')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
