## Autocorrelation of positional encoding

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
D = 512
base_frequency = 10000

# Only positive k for log scale
k = np.arange(0, 10001)

def compute_autocorrelation(k_vec, D_val, base_freq):
    i = np.arange(D_val // 2)
    omega = 1 / (base_freq ** (2 * i / D_val))
    cos_terms = np.cos(np.outer(k_vec, omega))
    return (2 / D_val) * np.sum(cos_terms, axis=1)

# Compute
R = compute_autocorrelation(k, D, base_frequency)

# --- Plot ---
plt.figure(figsize=(10, 5))
plt.plot(k, R, linewidth=2)

plt.xscale('log')  # key
plt.xlabel("Relative Distance (k) [log scale]")
plt.ylabel("Relative Position Similarity  (P_n · P_{n+k})")
plt.title(f"Autocorrelation of Positional Encoding (D={D}, base_freq={base_frequency})")

plt.grid(True, which='both', linestyle='--', alpha=0.4)
plt.xlim(1, 10000)
plt.tight_layout()
plt.show()
```
