# Sigmoid and Derivative

```python
import numpy as np
import matplotlib.pyplot as plt

# Define sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Derivative of sigmoid
def sigmoid_derivative(z):
    return z * (1 - z)

# Generate input values
z = np.linspace(-6, 6, 400)
sig = sigmoid(z)
sig_der = sigmoid_derivative(sig)

# Plot
plt.figure(figsize=(8, 5))
plt.plot(z, sig, label='Sigmoid $\sigma(z)$', color='blue')
plt.plot(z, sig_der, label="Derivative $\sigma'(z)$", color='red', linestyle='--')
plt.title('Sigmoid Function and Its Derivative')
plt.xlabel('z')
plt.ylabel('Value')
plt.grid(True)
plt.legend()
plt.tight_layout()
```
