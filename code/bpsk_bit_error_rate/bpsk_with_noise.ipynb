### BPSK Modulation

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import erfc

# Parameters
N = int(1e6)  # number of bits or symbols
np.random.seed(100)  # initializing the random number generator

# Eb/N0 values
Eb_N0_dB = np.arange(-3, 11)  # multiple Eb/N0 values

# Transmitter
ip = np.random.rand(N) > 0.5  # generating 0,1 with equal probability
s = 2 * ip - 1  # BPSK modulation: 0 -> -1; 1 -> 1

# Generate noise for SNR = 6 dB
Eb_N0 = 6  # SNR in dB
n = (1 / np.sqrt(2)) * (np.random.randn(N) + 1j * np.random.randn(N))  # AWGN, 0dB variance
y = s + 10 ** (-Eb_N0 / 20) * n  # additive white Gaussian noise
```

### Scatter plot - BPSK with Eb/N0=6dB

```python

# Plot BPSK symbols and received noisy symbols
plt.figure(figsize=(8, 6))
plt.scatter(np.real(s[:1000]), np.imag(s[:1000]), color='blue', label='Transmitted Symbols (s)', alpha=0.5)
plt.scatter(np.real(y[:1000]), np.imag(y[:1000]), color='red', label='Received Symbols (s + noise)', alpha=0.5)
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend()
plt.title('BPSK Symbols and Received Noisy Symbols (SNR = 6 dB)')
plt.xlabel('In-Phase')
plt.ylabel('Quadrature')
plt.show()
```

### Probability Distribuition of BPSK with Eb/N0 = 6dB


```python
# Plot PDF of the received symbols
plt.figure(figsize=(8, 6))
plt.hist(np.real(y), bins=100, density=True, color='blue', alpha=0.6, label='PDF of Received Symbols (Real Part)')
plt.axvline(x=-1, color='red', linestyle='--', label='Ideal Symbol (-1)')
plt.axvline(x=1, color='green', linestyle='--', label='Ideal Symbol (1)')
plt.grid(True, linestyle='--', linewidth=0.5)
plt.legend()
plt.title('PDF of the Received Symbols (Real Part) at SNR = 6 dB')
plt.xlabel('Real Part of Received Symbols')
plt.ylabel('Probability Density')
plt.show()
```
